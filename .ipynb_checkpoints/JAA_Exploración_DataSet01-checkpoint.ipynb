```python
# Define los nombres de las columnas y sus tipos de datos
headers = ['date', 'time', 'door_state', 'sphone_signal', 'label', 'type']

# Define las columnas que deben ser tratadas como fechas y tiempo
parse_dates = {'datetime': ['date', 'time']}

# Define los valores que deben ser considerados como nulos durante la importación
na_values = ['', 'NA', 'null', 'NaN', 'nan']  # Agrega aquí cualquier otro valor que pueda representar un nulo

# Lee el archivo CSV especificando los parámetros
Garage_Door = pd.read_csv(file_path2, header=0, names=headers, parse_dates=parse_dates, na_values=na_values)
Garage_Door
```

```python
Garage_Door.info()
```

```python
Garage_Door.door_state.unique()
```

```python
Garage_Door.sphone_signal.unique()
```

```python
# Elimina las filas que contienen valores nulos en las columnas booleanas
Garage_Door = Garage_Door.dropna(subset=['door_state','sphone_signal'])
Garage_Door
```

```python
Garage_Door.door_state.unique()
```

```python
Garage_Door.sphone_signal.unique()
```

```python
# Eliminar los espacios en blanco alrededor de las cadenas en la columna sphone_signal
Garage_Door.loc[:,'sphone_signal'] = Garage_Door['sphone_signal'].str.strip()

# Reemplazar las cadenas 'true', 'false', '1' y '0' por los valores booleanos correspondientes
Garage_Door.loc[:,'sphone_signal'] = Garage_Door['sphone_signal'].replace({'true': True, 'false': False, '1': True, '0': False})

# Convertir cualquier valor numérico 0.0 o 1.0 a booleano
Garage_Door.loc[:,'sphone_signal'] = Garage_Door['sphone_signal'].astype(bool)

# Ahora la columna sphone_signal debe contener solo valores booleanos True y False
Garage_Door.sphone_signal.unique()
```

```python
Garage_Door.info()
```

```python
# Resetear el índice
Garage_Door.reset_index(drop=True, inplace=True)
Garage_Door
```

```python
Garage_Door.type.unique()
```

```python
Garage_Door.label.unique()
```

```python
new_Garage_Door = Garage_Door.copy()
new_Garage_Door['label'] = new_Garage_Door['label'].astype('int8')
```

```python
new_Garage_Door.loc[:,'label'] = new_Garage_Door['label'].astype('int8')
```

```python
new_Garage_Door.loc[:,'datetime'] = pd.to_datetime(new_Garage_Door['datetime'])
```

```python
new_Garage_Door.loc[:,'door_state'] = new_Garage_Door['door_state'].astype(str)
# Convertir la columna type a tipo de datos category
new_Garage_Door['door_state'] = new_Garage_Door['door_state'].astype('category')
```

```python
new_Garage_Door.info()
```

```python
new_Garage_Door.loc[:,'type'] = new_Garage_Door['type'].astype(str)
# Convertir la columna type a tipo de datos category
new_Garage_Door['type'] = new_Garage_Door['type'].astype('category')
```

```python
Garage_Door = new_Garage_Door.copy()
Garage_Door.info()
```

# Combinando todos los datasets

```python
for df in [Fridge, Garage_Door, GPS_Tracker, Modbus, Motion_Light, Thermostat]:
    df.dropna(inplace=True)
    df.info()
```

```python
for df in [Fridge, Garage_Door, GPS_Tracker, Modbus, Motion_Light, Thermostat]:
    df['datetime'] = pd.to_datetime(df['datetime']).dt.floor('min').dt.strftime('%Y-%m-%d %H:%M')

# Esto convertirá la columna 'datetime' a formato de fecha y hora, truncando los segundos, y luego formateará la columna en el formato deseado.
```
